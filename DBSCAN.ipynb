```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
moons = pd.read_csv('cluster_moons.csv')
circles =pd.read_csv('cluster_circles.csv')
blobs = pd.read_csv('cluster_blobs.csv')
```

```python
plt.figure(figsize=(8,5),dpi=200)
sns.scatterplot(data=moons,x='X1',y='X2')
plt.figure(figsize=(8,5),dpi=200)
sns.scatterplot(data=circles,x='X1',y='X2')
plt.figure(figsize=(8,5),dpi=200)
sns.scatterplot(data=blobs,x='X1',y='X2')
```

```python
from sklearn.cluster import KMeans, DBSCAN
def model_realize(model,data):
    labels = model.fit_predict(data)
    plt.figure(figsize=(8,5),dpi=200)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels)
```

```python
model_realize(KMeans(n_clusters=2),moons)
```

```python
model_realize(KMeans(n_clusters=2),circles)
```

```python
model_realize(KMeans(n_clusters=3),blobs)
```

```python
model_realize(DBSCAN(eps=0.2),moons)
```

```python
model_realize(DBSCAN(eps=1.5),blobs)
```

```python
model_realize(DBSCAN(eps=0.2),circles)
```
